# Write a Python class named Circle constructed by a radius and two methods which will compute the area and the perimeter of a circle

```python
class Circle():
    def __init__(self, radius):
        self.radius = radius
    def ComputeArea(self):
        print(str(3.14*self.radius**2))
    def ComputePerimeter(self):
        print(str(2*3.14*self.radius))
```

```python
c = Circle(4)
c.ComputeArea()
c.ComputePerimeter()
```

# Write a Python class which has two methods get_String and print_String. get_String accept a string from the user and print_String print the string in upper case.

```python
class GetPrintString():
    def get_String(self):
        self.strValue = input("Please input string: ")
    def print_String(self):
        print(self.strValue)
```

```python
g = GetPrintString()
g.get_String()
g.print_String()
```

# Write a Python class to reverse a string word by word. - Go to the editor Input string : 'hello .py' Expected Output : '.py hello'

```python
class StringManipulation():
    def __init__(self, strValue):
        self.strValue = strValue
    def reverseString(self):
        newStringArray = self.strValue.split(" ")
        newStringArray.reverse()
        self.strValue = ""
        for eachWord in newStringArray:
            self.strValue += " " + eachWord
        self.strValue = self.strValue.strip()
```

```python
s = StringManipulation("A quick brown fox jumped over the lazy dog.")
s.reverseString()
s.strValue
```
