# 1. Calculate area of circle.

```python
def calcArea(radius, pi = 3.14):
    area = pi * radius**2
    return f"Area of circle is: {area}"
```

```python
radius = float(input("Please enter radius: "))
calcArea(radius)
```

# 2. Write function to check if a value exists in a list

```python
def checkVal(list_to_check, value_to_check):
    for each_value in list_to_check:
        if each_value == value_to_check:
            return f"{value_to_check} found in the list."
    return f"{value_to_check} not found in the list."
```

```python
data = ["python", "r", "java", "pascal", "c", "javascript", "assembley", "html", "c++"]
value_to_check = input("Please enter value to check in the list: ")
checkVal(data, value_to_check)
```

# 3. Write a function to shift list item to right n times

```python
def shiftItems(ListName, n):
    for i in range(n):
        ListName.insert(0, ListName.pop())
    return ListName
```

```python
data = ["python", "r", "java", "pascal", "c", "javascript", "assembley", "html", "c++"]
n = int(input("Please enter number of times to shift list: "))
shiftItems(data, n)
```

# 5. Function with **kwargs:

```python
def totalAndPercentage(subject_marks):
    total = 0
    percentage = 0
    for subject_mark in subject_marks.values():
        total += subject_mark
    percentage = total / len(subject_marks)
    return total, percentage

def minMaxMarks(subject_marks, option): #option 1=Min, 2=Max
    subjectsList = [k for k in subject_marks]
    subject = subjectsList[0]
    marks = subject_marks[subject]
    if option == 1:
        for key, value in subject_marks.items():
            if value < marks:
                subject = key
        return subject
    else:
        for key, value in subject_marks.items():
            if value > marks:
                marks = value
                subject = key
        return subject

def showDetails(**kwargs):
    student_details = ""
    student_details += "Hello " +  kwargs["name"] + "\n"
    student_details += "Here is your result for class of " + kwargs["classs"] + ".\n"
    student_details += "Your score card is as follows.\n"
    student_details += "ClassMarks:\n"
    for subject_name, marks in kwargs["marks"].items():
        student_details += "\t" + subject_name + " : " + str(marks) + "\n"
    total, percentage = totalAndPercentage(kwargs["marks"])
    student_details += "Total marks are: " + str(total) + "\n"
    student_details += "Percentage is as follows: " + str(percentage) + "%\n"
    student_details += "Maximum marks are in \"" + minMaxMarks(kwargs["marks"], 2).title() + "\"\n"
    student_details += "Minimum marks are in \"" + minMaxMarks(kwargs["marks"], 1).title() + "\"\n"
    if kwargs["nextClass"] == True:
        student_details += "You are promoted to next class."
    return student_details
```

```python
print(showDetails(name = "Ali", classs = "AI", marks = {"math" : 15, "physics" : 90, "biology" : 70, "computer" : 50}, date = "1 Feb 2020", nextClass = True))
```
